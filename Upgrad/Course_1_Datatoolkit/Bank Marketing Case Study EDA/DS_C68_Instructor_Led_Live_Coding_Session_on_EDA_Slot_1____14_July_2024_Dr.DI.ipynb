### What is EDA (Exploratory Data Analysis)?

Exploratory Data Analysis (EDA) is an approach to analyzing data sets to summarize their main characteristics, often using visual methods. It is an essential step in the data analysis process that involves investigating the data to discover patterns, spot anomalies, test hypotheses, and check assumptions with the help of summary statistics and graphical representations.

### Key Steps in EDA

1. **Data Collection**: Gathering relevant data from various sources.
2. **Data Cleaning**: Handling missing values, outliers, and inconsistent data.
3. **Data Transformation**: Normalizing and scaling data, creating new features, and encoding categorical variables.
4. **Data Visualization**: Using charts and graphs to visualize distributions, relationships, and trends.
5. **Summary Statistics**: Calculating mean, median, mode, standard deviation, and other statistical metrics.
6. **Hypothesis Testing**: Formulating and testing hypotheses to draw insights.

### Industry Significance of EDA

EDA is crucial in various industries for making data-driven decisions. Here are some real-time examples:

1. **Finance**:
   - **Fraud Detection**: Banks and financial institutions use EDA to analyze transaction data and detect fraudulent activities. For example, by visualizing transaction patterns and outliers, analysts can identify unusual spending behaviors indicative of fraud.
   - **Risk Management**: EDA helps in assessing credit risk by examining historical data on loans and defaults, identifying key factors that influence risk, and developing predictive models.

2. **Healthcare**:
   - **Patient Data Analysis**: Hospitals and clinics use EDA to analyze patient records, identify trends in disease outbreaks, and improve treatment plans. For instance, visualizing patient demographics and treatment outcomes can reveal insights into the effectiveness of different therapies.
   - **Medical Research**: Researchers use EDA to explore clinical trial data, uncover patterns in patient responses, and refine research hypotheses.

3. **Retail**:
   - **Customer Segmentation**: Retailers use EDA to analyze customer purchase data, segment customers based on buying behavior, and tailor marketing strategies. For example, clustering analysis can help identify distinct customer groups with similar preferences.
   - **Inventory Management**: EDA assists in examining sales data to predict demand, optimize inventory levels, and reduce stockouts or overstock situations.

4. **Marketing**:
   - **Campaign Analysis**: Marketers use EDA to evaluate the performance of marketing campaigns by analyzing response rates, conversion rates, and ROI. Visualization tools like heatmaps can show which segments responded best to different campaigns.
   - **Customer Feedback**: EDA helps analyze customer reviews and feedback to identify common issues, sentiment trends, and areas for improvement.

5. **Manufacturing**:
   - **Quality Control**: Manufacturers use EDA to monitor production data, identify defects, and improve product quality. For example, control charts can help visualize the stability of manufacturing processes and detect variations.
   - **Predictive Maintenance**: EDA is used to analyze machine sensor data to predict equipment failures and schedule maintenance before breakdowns occur, thus reducing downtime and maintenance costs.

### Example: EDA in Retail

**Objective**: Improve Sales through Customer Segmentation

**Steps**:
1. **Data Collection**: Gather transaction data from point-of-sale systems, including customer demographics, purchase history, and product details.
2. **Data Cleaning**: Handle missing values and outliers in the data.
3. **Data Transformation**: Create new features like total spending, average purchase value, and frequency of purchases.
4. **Data Visualization**: Use scatter plots to visualize relationships between variables, histograms to show the distribution of spending, and box plots to identify outliers.
5. **Summary Statistics**: Calculate mean and median purchase values, standard deviation of spending, and other key metrics.
6. **Hypothesis Testing**: Test hypotheses such as "High-frequency customers spend more on average" using statistical tests.

**Outcome**:
- Identify distinct customer segments (e.g., high spenders, frequent shoppers) using clustering analysis.
- Tailor marketing strategies for each segment, such as offering loyalty rewards to high-frequency customers and personalized discounts to high spenders.
- Visualize customer segments using bar charts and pie charts to communicate insights to stakeholders.

### Conclusion

EDA is a foundational step in data analysis that enables organizations to understand their data deeply and make informed decisions. By uncovering hidden patterns and relationships, EDA helps businesses across various industries optimize operations, improve customer experiences, and gain a competitive edge.

### Strategies for filling in the missing values

1. Categorical columns.

Fill the missing values with the mode, ffill, bfill, constant. What you use for filling in completely depends on the context of the business problem.

2. Numerical columns.

Check for the distribution of the column, using a histogram.

If the histogram is normally distributed, fill the missing values by the mean of that column.

If the histogram is not normally distributed, then there is a possibility of outliers in the data. In such scenario, fill the missing values by the median of the data. Again to remind, what value you fill depends on the context of the business problem.

Other than this, we also have some advanced imputation techniques which are very popularly used in the industry, namely SimpleImputer, KNNImputer, MICE.

```python
# Fill the missing values in the Embarked column by mode

df.Embarked.fillna(df.Embarked.mode()[0], inplace=True)
```

```python
# Lets find the missing value count in each column in descending order in percentage

((df.isnull().sum().sort_values(ascending=False)) / df.shape[0]) * 100
```

```python
# Plot histogram of Age column

sns.distplot(df.Age)
plt.axvline(df.Age.mean(), color='red', linestyle="dashed")
plt.axvline(df.Age.median(), color='green', linestyle="dashed")
plt.show()
```

```python
df.Age.skew()
```

Rule of thumb:

>For skewness values between -0.5 and 0.5, the data exhibit approximate symmetry.

>Skewness values within the range of -1 and -0.5 (negative skewed) or 0.5 and 1(positive skewed) indicate slightly skewed data distributions.

>Data with skewness values less than -1 (negative skewed) or greater than 1 (positive skewed) are considered highly skewed.

```python
# Fill Age column missing values with mean of Age column

df.Age.fillna(df.Age.mean(), inplace=True)
```
